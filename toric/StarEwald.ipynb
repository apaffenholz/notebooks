## Counterexamples to the Star Ewald Conjecture.

For background see 

Dusa McDuff, Displacing Lagrangian toric fibers via probes, arxiv:0904.1686, [[doi]](https://doi.org/10.48550/arXiv.0904.1686)

```polymake
$polydb = polyDB();
$collection = $polydb->get_collection("Polytopes.Lattice.SmoothReflexive");
```

```polymake
sub star_ewald { 
  my $p=shift; 
  my $se=1; my $nf=$p->N_FACETS; 
  my $points_on_facets = new Array<Set<Int>>($nf); 
  my $opposite_points_on_facets = new Array<Set<Int>>($nf); 
  my $inter=intersection($p,scale($p,-1)); 
  my $lattice_points=$inter->LATTICE_POINTS;

  foreach my $lp (0..$lattice_points->rows-1) { 
    my $l_neg = -$lattice_points->[$lp]; $l_neg->[0]=1; 
    foreach my $f (0..$nf-1) { 
      if ( $p->FACETS->[$f]*$l_neg == 0 ) {
        $opposite_points_on_facets->[$f]+=$lp; 
      } 
      if ( $p->FACETS->[$f]*$lattice_points->[$lp] == 0 ) {
        $points_on_facets->[$f]+=$lp; 
      } 
    } 
  }

  my $hd = polarize($p,no_coordinates=>1)->HASSE_DIAGRAM; 
  foreach my $dim (0..$p->DIM-1) { 
    foreach my $s (@{$hd->nodes_of_rank($dim+1)}) { 
      my $delstar=new Set<Int>; 
      my $opp_star=new Set<Int>; 
      my $mult_points = new Set<Int>; 
      foreach my $f (@{$hd->FACES->[$s]}) { 
        $mult_points += $delstar * $points_on_facets->[$f]; 
        $delstar += $points_on_facets->[$f]; 
        $opp_star += $opposite_points_on_facets->[$f]; 
      } 
      $delstar -= $mult_points; 
      if ( ($delstar-$opp_star)->size==0 ) { 
        print "condition violated on intersection of facets: ", $hd->FACES->[$s], "\n"; 
    	print "star:\n ", $opp_star, "\n";
    	print "delstar:\n ", $delstar, "\n";
        return false;
      } 
    } 
  } 
  return true; 
}
```

### The examples in dimension 6

```polymake
$p61=$collection->find_one({"_id"=>"F.6D.0070"});
$p62=$collection->find_one({"_id"=>"F.6D.0071"});
$p63=$collection->find_one({"_id"=>"F.6D.2231"});
```
